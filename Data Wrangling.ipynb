```python
import pandas as pd
import numpy as np
```

Dataframes that were created in Data Collection file were saved as csv files and have now been added back into this new file for further cleaning and wrangling. 

```python
hurts_df = pd.read_csv('/Users/noah/Desktop/Player Efficiency Analysis/CSV/hurts_dataset.csv')
mahomes_df = pd.read_csv('/Users/noah/Desktop/Player Efficiency Analysis/CSV/mahomes_dataset.csv')
burrow_df = pd.read_csv('/Users/noah/Desktop/Player Efficiency Analysis/CSV/burrow_dataset.csv')
carr_df = pd.read_csv('/Users/noah/Desktop/Player Efficiency Analysis/CSV/carr_dataset.csv')
```

Checking the Status of the players, it is found that the only reliable stat of comparison is when the player Starts a game. Therefore, other values such as DNS or Inactive must be removed to ensure a fair comparison. 

```python
hurts_df.value_counts('Status')
```

```python
hurts_df.drop(hurts_df[hurts_df['Status'] == 'Inactive'].index, inplace=True)
hurts_df.drop(hurts_df[hurts_df['Status'] == 'DNS'].index, inplace=True)


mahomes_df.drop(mahomes_df[mahomes_df['Status'] == 'Did Not Play'].index, inplace=True)
mahomes_df.drop(mahomes_df[mahomes_df['Status'] == 'Inactive'].index, inplace=True)


burrow_df.drop(burrow_df[burrow_df['Status'] == 'Did Not Play'].index, inplace=True)
burrow_df.drop(burrow_df[burrow_df['Status'] == 'Inactive'].index, inplace=True)
burrow_df.drop(burrow_df[burrow_df['Status'] == 'Injured Reserve'].index, inplace=True)


carr_df.drop(carr_df[carr_df['Status'] == 'Inactive'].index, inplace=True)
```

Since the Result column has string values representing the score and the result of the game, creating a new column with numerical values for the result is a must for creating ease of analysis. The column will be labelled as Class and will contain 3 possible values. 1 = Win, 0 = Loss, and 0.5 = Tie (the NFL considers a tie to be a half win/half lose).

```python
result_array = []
column = hurts_df['Result']
for i in column:
    if 'W' in i:
        result_array.append(1)
    elif 'L' in i:
        result_array.append(0)
    else:
        result_array.append(0.5)

hurts_df['Class'] = result_array

result_array = []
column = mahomes_df['Result']
for i in column:
    if 'W' in i:
        result_array.append(1)
    elif 'L' in i:
        result_array.append(0)
    else:
        result_array.append(0.5)

mahomes_df['Class'] = result_array

result_array = []
column = burrow_df['Result']
for i in column:
    if 'W' in i:
        result_array.append(1)
    elif 'L' in i:
        result_array.append(0)
    else:
        result_array.append(0.5)

burrow_df['Class'] = result_array

result_array = []
column = carr_df['Result']
for i in column:
    if 'W' in i:
        result_array.append(1)
    elif 'L' in i:
        result_array.append(0)
    else:
        result_array.append(0.5)

carr_df['Class'] = result_array
```
