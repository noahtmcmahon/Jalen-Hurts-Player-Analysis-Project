```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hurts_df = pd.read_csv('/Users/noah/Desktop/Player Efficiency Analysis/CSV/hurts_dataset2.csv')
mahomes_df = pd.read_csv('/Users/noah/Desktop/Player Efficiency Analysis/CSV/mahomes_dataset2.csv')
burrow_df = pd.read_csv('/Users/noah/Desktop/Player Efficiency Analysis/CSV/burrow_dataset2.csv')
carr_df = pd.read_csv('/Users/noah/Desktop/Player Efficiency Analysis/CSV/carr_dataset2.csv')
```

In this section, a comparison will be made between each of the 4 quarterbacks Jalen Hurts (subject), Patrick Mahomes (number 1 ranked quarterback), Joe Burrow (quarterback from the same draft class as Hurts), and Derek Carr (an overall average quarterback). We will see how these 4 players compare and contrast and find out where Jalen Hurts ranked amongst the competition. 

```python
hurts_df['Name'] = 'Hurts'
mahomes_df['Name'] = 'Mahomes'
burrow_df['Name'] = 'Burrow'
carr_df['Name'] = 'Carr'
dfs = [hurts_df, mahomes_df, burrow_df, carr_df]
df = pd.concat(dfs)
```

```python
sns.barplot(x='Name', y='Pass YDs', data=df, palette = ['#388E3C', '#FF321E', '#FF922B','#FFCC3B'], estimator=sum, errorbar=None)
plt.xlabel('Quarterbacks')
plt.ylabel('Passing Yards')
plt.title('Total Passing Yards by Quarterback')
plt.show()
```

Hurts has the lowest career passing yards of the four quarterbacks. While Carr and Mahomes have been in the league for a significantly longer time than Hurts, Burrow was in the same draft class and has nearly 4,000 more yards. Hurts may have other strengths that these quarterbacks do not. 

```python
colors = ['#388E3C', '#FF321E', '#FF922B','#FFCC3B']
sns.barplot(x='Name', y='Rush YDs', data=df, palette = colors , estimator=sum, errorbar=None)
plt.xlabel('Quarterbacks')
plt.ylabel('Rushing Yards')
plt.title('Total Rushing Yards by Quarterback')
plt.show()
```

Hurts has a larger total rushing yards than any other quarterback. So while his passing yards may be below what the other 3 quarterbacks are producing, his run game is much better. 

```python
fig, ax = plt.subplots(figsize=(12, 6))
sns.lineplot(x = 'Pass Att', y = 'Pass YDs', hue = 'Name', ax = ax, data = df, errorbar =None, palette = colors )
plt.xlabel('Passing Attempts')
plt.ylabel('Passing Yards')
plt.title('Passing Yards per Attempt')
plt.show()
```

Each of the four quarterbacks passing yards increase with an increase in attempts. This makes sense as the more attempts will normally lead to a great overall possibility of more yardage. The big difference comes with maximum passing attempts. Mahomes, Burrow, and Carr all have come close to or surpassed 60 attempts in a single game. Hurts has never broken 50. This shows an emphasis on passing for the other 3 quarterbacks that Hurts does not show. 

```python
fig, ax = plt.subplots(figsize=(12, 6))
sns.lineplot(x = 'Rush Att', y = 'Rush YDs', hue = 'Name', ax = ax, data = df, errorbar =None, palette = colors )
plt.xlabel('Rushing Attempts')
plt.ylabel('Rushing Yards')
plt.title('Rushing Yards per Attempt')
plt.show()
```

As expected, Hurts has significantly more rushing attempts and due to this, has significantly more rushing yards. Carr barely reaches a maximum of 7 attempts, while Burrow and Mahomes both max out between 10 to 12 attempts. Hurts has had games with 18 rushing attempts. The things that has Hurts standing out from his competition is his ability to run. 

```python
fig, ax = plt.subplots(figsize=(12, 6))
sns.lineplot(x = 'Year', y = 'Cmp%', hue = 'Name', ax = ax, data = df, errorbar =None, palette = colors )
plt.xlabel('Year')
plt.ylabel('Completion %')
plt.title('Completion % by Year')
plt.show()
```

Jalen Hurts has seen the most improvement in completion percentage over his NFL career jumping from between 50% and 60% his rookie year to 67% at his peak. The other three quarterbacks all started with better completion percentages but they have not alwasy trended upward. Carr increases greatly between 2014 and 2019 but since then he has been on a decline. Burrow started high in 2020 and improved through that year. But since 2021 he has been slowly declining. Mahomes has been the most consistent of any of the quarterbacks. Starting with a high completion percentage he hasn't been able to make much improvements. Hurts and Mahomes have nearly the same completion percentage as of 2022. So while Hurts may not throw the ball as much, he is just as consistent in 2022-2023 as the number 1 ranked quarterback.

```python
fig, ax = plt.subplots(figsize=(12, 6))
g = sns.barplot(x='Year', y='Class', hue = 'Name', data=df, ax = ax, palette = colors , estimator= np.mean, errorbar=None)
plt.xlabel('Year')
g.set_xticklabels(['2014','2015','2016','2017','2018', '2019', '2020', '2021', '2022', '2023'])
plt.ylabel('Win Percentage')
plt.title('Win Percentage by Year')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
g = sns.lineplot(x='Year', y='Class', hue = 'Name', data=df, ax = ax, palette = colors , estimator= np.mean, errorbar=None)
plt.xlabel('Year')
plt.xticks(np.arange(min(df['Year']), max(df['Year'])+1, 1))
g.set_xticklabels(['2014','2015','2016','2017','2018', '2019', '2020', '2021', '2022', '2023'])
plt.ylabel('Win Percentage')
plt.title('Win Percentage by Year')
plt.show()
```

The two graphs above show the same information but in two different formats. The bar graph helps to show each years comparison between the players. As we can see, Jalen Hurts has a higher win percentage in 2022 and 2023 than the other 3 quarterbacks. This shows great improvement to his game while the other three quarterbacks have all had a decrease in win percentage.

The line plot helps to show the progression of each player through the years of their career. Carr has peaked in 2016 while Jalen peaked in 2022. Burrow also peaked in 2022 but has a lower win percentage at his peak. Mahomes has had a superior win percentage to all other players in 2017 but has yet to match that. He only has one recorded game in 2017 however which skews that data.
