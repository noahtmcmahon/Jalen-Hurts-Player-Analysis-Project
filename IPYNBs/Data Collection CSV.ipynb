```python
import pandas as pd
import numpy as np
```

Created CSV files from Pro Football Reference website for 4 players. Added those files into jupyter using pandas. The 4 players being used include:
    1. Jalen Hurts (subject of analysis)
    2. Patrick Mahomes (number 1 quarterback in the NFL)
    3. Joe Burrow (same draft class as subject)
    4. Derek Carr (average quarterback for comparison)

```python
hurts_df = pd.read_csv('/Users/noah/Desktop/Player Efficiency Analysis/CSV/HurtsCareer.csv')
mahomes_df = pd.read_csv('/Users/noah/Desktop/Player Efficiency Analysis/CSV/MahomesCareer.csv')
burrow_df = pd.read_csv('/Users/noah/Desktop/Player Efficiency Analysis/CSV/BurrowCareer.csv')
carr_df = pd.read_csv('/Users/noah/Desktop/Player Efficiency Analysis/CSV/CarrCareer.csv')

pd.set_option('display.max_columns', None)
hurts_df.head()
```

Removing unneeded columns from the dataframe and rename columns to be more consistent with the statisitic. Renamed certain statistics to ensure ease of understanding during EDA phase. Changed dtype of numerical statisitics and changed date string to datetime type. Filled NaN values with the column average to ensure statisitics are not thrown off due to inaccurate values. 

These steps were done to all 4 players' dataframes.

```python
hurts_df.columns = hurts_df.iloc[0]
hurts_df = hurts_df.drop(0)
hurts_df.reset_index(drop=True, inplace=True)

list_of_columns = [0,1,2,4,6,7,8,9,10,12,13,14,15,16,18,22,23,34,48]
hurts_df = hurts_df.iloc[:,list_of_columns]
hurts_df.head()

new_names = ['Rank','Year','Date','Week','Team','H/A', 'Opp','Result','Status','Pass Att',
            'Cmp%','Pass YDs','Pass TDs','Int','Sacks','Rush Att','Rush YDs','TDs','Snaps']
old_names = hurts_df.columns
hurts_df.rename(columns=dict(zip(old_names, new_names)), inplace=True)

new_column_name = 'Pass Att'
hurts_df.columns.values[9] = new_column_name
new_column_name2 = 'Pass YDs'
hurts_df.columns.values[11] = new_column_name2
new_column_name3 = 'Pass TDs'
hurts_df.columns.values[12] = new_column_name3

hurts_df = hurts_df.drop(62)

hurts_df['H/A'] = hurts_df['H/A'].fillna('Home')
hurts_df['H/A'] = hurts_df['H/A'].replace('@', 'Away')
hurts_df['Status'] = hurts_df['Status'].fillna('DNS')
hurts_df['Status'] = hurts_df['Status'].replace('*', 'Start')

hurts_df.iloc[:,0:2] = hurts_df.iloc[:, 0:2].astype(float)
hurts_df.iloc[:,3:4] = hurts_df.iloc[:, 3:4].astype(float)
hurts_df['Date'] = pd.to_datetime(hurts_df['Date'])
hurts_df.iloc[:,9:] = hurts_df.iloc[:,9:].astype(float)

averaging_columns = ['Pass Att','Cmp%','Pass YDs','Pass TDs','Int','Sacks','Rush Att','Rush YDs','TDs','Snaps']
for i in averaging_columns:
    hurts_df[i] = hurts_df[i].fillna((hurts_df[i].mean()))
```
