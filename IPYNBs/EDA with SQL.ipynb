```python
sql select "Opp", SUM("Pass YDs") as "Total Pass YDs" from HURTSTBL group by "Opp" order by SUM("Pass YDs") DESC
```

This table shows the total number of passing yards by Opponent. Jalen has thrown the highest passing yards to Washington while he has thrown the lowest to Green Bay. The number of times each of these teams was played could be significantly biasing the numbers. 

```python
sql select "Opp", ROUND(AVG("Pass YDs")) as "Avg Pass Yards" from HURTSTBL group by "Opp" order by AVG("Pass YDs") DESC
```

Now we can see that Washington is closer to the middle and Jalen has probably played them more, skewing the total passing yards statisitic. Green Bay is still at the bottom and now Tennesse is at the top with 380 yards. 
NOTE: Both the top and bottom of this table are numbers from only one game played however. Not a long term average. 

```python
sql select "Opp", ROUND(AVG("Cmp%"),2) as 'Comp %' from HURTSTBL group by "Opp" order by AVG("Cmp%") DESC
```

This may be a better way of understanding Hurts's efficiency against a team. By looking at his completion percentage by team, we see how many of his passes were completed. From this table, we can gather that Hurts is most efficient against Minnesota and least efficient against Las Vegas. 

```python
sql select "Opp", ROUND(AVG("Rush YDs")) as "Avg Rush Yards" from HURTSTBL group by "Opp" order by AVG("Rush YDs") DESC
```

Due to the previous possible bias in the table for total passing yards, I decided not to do a total rushing yards table and skip to the averages. Hurts had the highest against Green Bay with 157 and the lowest against Pittsburgh with 10. 

```python
sql select "Opp", ROUND(AVG("Int"),1) as "Avg Ints", SUM("Int") as "Total Ints" from HURTSTBL group by "Opp" order by AVG("Int") DESC
```

Hurts has thrown the most interceptions to the Giants but on average he has thrown the most per game to the Jets. 

```python
sql select "Opp", ROUND(AVG("Sacks"),1) as "Avg Sacks", SUM("Sacks") as "Total Sacks" from HURTSTBL group by "Opp" order by AVG("Sacks") DESC
```

Hurts has been sacked the most by Washington but on average he has been sacked the most per game by Kansas City. Houston, and Arizona.

```python
sql select "H/A", "Class" as "Result", COUNT("Class") as "Win/Loss Count" from HURTSTBL group by "H/A", "Class"
```

This table shows Hurts's wins and losses based on whether the game was at Home or Away. He wins slightly more games Away but the numbers are almost identical. 

Something that stood out to me immediately from the above tables was that Green Bay appeared at the top of the rushing yards table and at the bottom of the passing yards table. Without looking into the tables beyond this, my hypothesis is the more rushing yards Hurts has the less passing yards he has or vice versa. Now I will find out if this is true. 

```python
sql select ROUND(AVG("Pass YDs")) as "Over/Under AVG", Count("Pass YDs") as "Games Count" from HURTSTBL group by "Pass YDs" < (select AVG("Pass YDs") from HURTSTBL)
```

Out of the 46 games that we are using, 22 of those games he has thrown above his average in passing yards and 24 of those games he has thrown under his average in passing yards. 

```python
sql select ROUND(AVG("Rush YDs")) as "Over/Under AVG", Count("Rush YDs") as "Games Count" from HURTSTBL where "Pass YDs" > (select AVG("Pass YDs") from HURTSTBL) group by "Rush YDs" < (select AVG("Rush YDs") from HURTSTBL)
```

Out of those 22 games where he threw over his average, 13 of them he has run under his average in rushing yards. This shows that when he is able to throw for more yards, he tends to run the ball less. Or when he can't run the ball very well in game, he tends to throw the ball more. 

```python
sql select ROUND(AVG("Rush YDs")) as "Over/Under AVG", Count("Rush YDs") as "Games Count" from HURTSTBL where "Pass YDs" < (select AVG("Pass YDs") from HURTSTBL) group by "Rush YDs" < (select AVG("Rush YDs") from HURTSTBL)
```

When hurts passing yards are below his average, we can see that the hypothesis stays true. His rushing averages both went up. This means that the more passing yards he is getting, the less rushing yards he is inclined to produce and vice versa. But which is better? Which produces more wins? Rushing more and passing less or passing more and rushing less?

```python
sql select "Class" as "Result", COUNT("Class") as "Total" from HURTSTBL where "Pass YDs" > (select AVG("Pass YDs") from HURTSTBL) and "Rush YDs" < (select AVG("Rush YDs") from HURTSTBL) group by "Class"
```

Wins and Loses for when Hurts has above average passing yards and below average rushing yards. Hurts wins 69.2% of the time in this scenario. 

```python
sql select "Class" as "Result", COUNT("Class") as "Total" from HURTSTBL where "Pass YDs" < (select AVG("Pass YDs") from HURTSTBL) and "Rush YDs" > (select AVG("Rush YDs") from HURTSTBL) group by "Class"
```

Wins and Loses for when Hurts has below average passing yards and above average rushing yards. Hurts wins 72.7% of the time in this scenario. 

```python
sql select "Class" as "Result", COUNT("Class") as "Total" from HURTSTBL where "Pass YDs" > (select AVG("Pass YDs") from HURTSTBL) and "Rush YDs" > (select AVG("Rush YDs") from HURTSTBL) group by "Class"
```

Wins and Loses for when Hurts has above average passing yards and above average rushing yards. Hurts wins 66.7% of the time in this scenario. 

```python
sql select "Class" as "Result", COUNT("Class") as "Total" from HURTSTBL where "Pass YDs" < (select AVG("Pass YDs") from HURTSTBL) and "Rush YDs" < (select AVG("Rush YDs") from HURTSTBL) group by "Class"
```

Wins and Loses for when Hurts has below average passing yards and below average rushing yards. Hurts wins 76.9% of the time in this scenario. 

From these findings, it can be concluded that less passing yards seems to be the most productive play style for Jalen Hurts based off of the end goal of winning games. Rushing yards has some say in the matter in that when he rushes less he tends to win more. Hurts loses most frequently when he is both passing for above average yards and rushing for above average yards.
